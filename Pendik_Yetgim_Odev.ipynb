```python
#Yeni Sütun Ekleme
```

```python
df['Temperature_F'] = (df['Temperature'] * 9/5) + 32
```

```python
df[['Temperature', 'Temperature_F']].head()
```

```python
df['FeelsLike'] = df['Temperature'] - (df['Humidity'] / 100)
```

```python
df['FeelsLike'].head()
```

```python
#Gruplama
```

```python
df.groupby('City').size()
```

```python
ortalama_sicaklik = df.groupby('City')['Temperature'].mean()
ortalama_sicaklik
```
